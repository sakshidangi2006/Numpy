Series is an one dimensional array with any type of data

Creation of Series-:
There are multiple ways to create a Series

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
arr = np.array([1,2,3])
my_list = [11,22,33]
dict = {1:2, 2: 3, 3:4}
```

```python
pd.Series(my_list)  #default labels Series
```

pd.Series(my_list, index = labels)  #series created by list

```python
pd.Series(arr, index = labels) #Series created by array
```

```python
pd.Series(dict)  #Series created by dictionary
```

Data Frames:- it is a collection of multiple series

Creation of Data Frames:-

```python
labels = [1,2,3,4]
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data, index = labels)
df
```

```python
data = [
    ['John',28,'New York',65000],
    ['Anna',34,'Paris', 70000],
    ['Peter',29,'Berlin',62000],
    ['Linda',42,'London',85000]
]
columns = ['Name' , 'Age', 'City', 'Salary']
df2 = pd.DataFrame(data , columns = columns)
labels = [1,2,3,4]
df2
```

Selection and Indexing of Columns

```python
df2['Name']  #single column selection
```

```python
df2[['Name', 'City']] #selection of multiple columns
```

Creation of New Column

```python
df2['Designation'] = ['Doctor','Engineer','Artist','Engineer']
df2
```

Removing Columns

```python
#df2.drop('Designation', axis = 1)  #drop main data se remove nhi karta hai bas ek idea deta hai ki aapka data kaisa dikhega remove karne ke baad
```

```python
df2.drop('Designation', axis = 1 , inplace = True)
df2
```

```python
df2.drop(['City','Salary'], axis = 1) #multiple columns's removal
```

```python
df2
```

```python
df2.drop(0, axis = 0) # Removal of row
```

Selecting Rows

```python
df2.loc[0] # One row selection  or df2.iloc[0]
```

```python
df2.loc[[0,1]] #multiple row selection  
```

Selecting Subsets of Rows and Columns

```python
df2.loc[[0,1]][['City','Salary']]
```

```python
df2.loc[[2,3]][['Name','Age']]
```

Conditional Selection

```python
#I only want to see those people whose age is above 30
df2[df2['Age'] > 30 ]
```

```python
#I only want to see those people whose age is above 30 and from paris
df2[(df2['Age'] > 30) & (df2['City'] == 'Paris')]
```
