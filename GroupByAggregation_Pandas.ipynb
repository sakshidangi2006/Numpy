```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category':['A','B','A','B','A','B','A','B'],
    'Store': ['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales':[100,200,150,250,120,100,200,300],
    'Quantity':[10,15,12,18,8,20,15,25],
    'Date': pd.date_range('2023-01-01', periods = 8)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#Group by category and calculate the sum of sales
#******groupby create an object not dataFrames******
df.groupby('Category')['Sales'].sum()
```

```python
#Group by stores and calculate the sum of sales
df.groupby('Store')['Sales'].sum()
```

```python
#Group by multiple columns
#Group by category and store
df.groupby(['Category','Store'])['Sales'].sum()
```

Aggregation

```python
#mean median max min sum count standard deviation#
df['Sales'].mean()
```

```python
df['Sales'].agg(['mean','median','max','min','count','std','sum'])
```
