```python
#Which country has the highest population
highest_Population = df[df['population']== df['population'].max()]
highest_Population['country']
```

```python
#What is the capital of the country with highest population
highest_Population['capital_city']
```

```python
#Which country has the least population
least_Population = df[df['population'] == df['population'].min()]
least_Population['country']
```

```python
#What is the capital of country with least population
least_Population['capital_city']
```

```python
#Who is the leader of secondary highest populated country
df[df['population'] == df['population'].nlargest(2).iloc[1]]['political_leader']
```

```python
#How many countries are there whose political leaders are unknown
df[df['political_leader'].isna()]['country'].count()
```

```python
#How many countries have Republic in their full name
count = 0
def counting(txt):
    global count
    if 'republic' in txt.lower():
        count += 1
    return txt

df['country_long'] = df['country_long'].apply(counting)
print(count)
```

```python
#Which country in African region has highest population
africa_df =  df[df['continent'] == 'Africa']
africa_df[africa_df['population'] == africa_df['population'].max()]['country']
```
