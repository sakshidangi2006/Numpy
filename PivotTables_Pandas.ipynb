```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Date": pd.date_range('2023-01-01',periods = 20),
    "product": ['A','B','C','D'] *5,
    "region": ["North","West", "South","East","North",
    "West", "South","East","North","West", "South","East",
    "North","West", "South","East","North","West", "South","East"
    ],
    "sales": np.random.randint(100,1000,20),
    "units": np.random.randint(10,100,20),
    "rep": ["Alice","Bob","Carol","David","Emma","Frank","Grace","Henry","Isa",
    "Jack","Kath","Liam","Mia","Noa","Oly","Paul","Quinn","Ryan","Sopi","Tom"]
}
df = pd.DataFrame(data)
df['Month'] = df['Date'].dt.month_name()
df["Quarter"] = 'Q' + df["Date"].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df, values = 'sales', index = 'region', columns = 'product', aggfunc = 'median')
```

```python
pivot2 = pd.pivot_table(df, values = ['sales','units'], index = 'region', columns = 'product', aggfunc = 'median')
pivot2
```

```python
#Cross tabs are similar to pivot table but it is used for counting purpose not aggregation
pd.crosstab(df['product'],df['region'])
```
