```python
import numpy as np
import pandas as pd
```

Merging 2 DataFrames

```python
employees = pd.DataFrame({
    'employee_id': [1,2,3,4,5],
    'name': ['John','Anna','Peter','Linda','Bob'],
    'department': ['HR','IT','Finance','IT','HR']
})

salaries = pd.DataFrame({
    'employee_id': [1,2,3,9,10],
    'salary': [60000,65000,75000,80000,85000],
    'bonus': [5000,1000,7000,8000,12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'inner')
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'outer')
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'left')
```

```python
pd.merge(employees, salaries, on = 'employee_id', how = 'right')
```

Concatination of 2 DataFrames

```python
df1 = pd.DataFrame({
    'A': ['A0','A1','A2'],
    'B': ['B0','B1','B2'],
    'C': ['C0','C1','C2']
})

df2 = pd.DataFrame({
    'A': ['A3','A4','A5'],
    'B': ['B3','B4','B5'],
    'C': ['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) #concatination based on columns
```

```python
pd.concat([df1,df2], axis = 1) #concatination based on rows
```

Joining 2 DataFrames

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
},index = [1,2,3])

df2 = pd.DataFrame({
    'score': [85,90,75]
},index = [1,2,3])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```
